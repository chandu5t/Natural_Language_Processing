```python
import re
import string
import numpy as np
import pandas as pd
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer

# Download NLTK resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

print("All libraries imported successfully!")
```

## Step 1: Sample Dataset
We use a small labeled text dataset with categories (Positive, Negative, Neutral).

```python
# Sample dataset
data = {
    'text': [
        "The movie was absolutely wonderful and amazing!",
        "I hated the food at the restaurant. It was terrible.",
        "The weather today is okay, nothing special.",
        "She loves reading books and learning new things.",
        "This product is the worst I've ever bought!!!",
        "The service was decent, could have been better.",
        "Fantastic experience! Highly recommend to everyone.",
        "Not bad, but not great either. Just average.",
        "I absolutely love this new phone, it's incredible!",
        "The trip was boring and a complete waste of time."
    ],
    'label': [
        'Positive', 'Negative', 'Neutral',
        'Positive', 'Negative', 'Neutral',
        'Positive', 'Neutral', 'Positive', 'Negative'
    ]
}

df = pd.DataFrame(data)
print("Original Dataset:")
print(df)
```

## Step 2: Text Cleaning
- Convert to lowercase
- Remove punctuation
- Remove numbers
- Remove extra whitespaces

```python
def clean_text(text):
    """Perform text cleaning."""
    # 1. Lowercase
    text = text.lower()
    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)
    # 4. Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # 5. Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['cleaned_text'] = df['text'].apply(clean_text)

print("After Text Cleaning:")
print(df[['text', 'cleaned_text']])
```

## Step 3: Remove Stop Words
Using NLTK's English stop words list.

```python
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    """Remove stop words from text."""
    tokens = word_tokenize(text)
    filtered = [word for word in tokens if word not in stop_words]
    return ' '.join(filtered)

df['no_stopwords_text'] = df['cleaned_text'].apply(remove_stopwords)

print("After Stop Word Removal:")
print(df[['cleaned_text', 'no_stopwords_text']])
```

## Step 4: Lemmatization
Using NLTK's `WordNetLemmatizer` to reduce words to their base/dictionary form.

```python
lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    """Lemmatize each word in the text."""
    tokens = word_tokenize(text)
    lemmatized = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(lemmatized)

df['lemmatized_text'] = df['no_stopwords_text'].apply(lemmatize_text)

print("After Lemmatization:")
print(df[['no_stopwords_text', 'lemmatized_text']])
```

## Step 5: Label Encoding
Convert text labels (Positive, Negative, Neutral) into numeric values using `LabelEncoder`.

```python
le = LabelEncoder()
df['encoded_label'] = le.fit_transform(df['label'])

print("Label Encoding Mapping:")
for cls, enc in zip(le.classes_, le.transform(le.classes_)):
    print(f"  {cls} --> {enc}")

print("\nDataset with Encoded Labels:")
print(df[['label', 'encoded_label']])
```

## Step 6: TF-IDF Representation
Create TF-IDF feature vectors from the preprocessed (lemmatized) text.

```python
# Initialize TF-IDF Vectorizer
tfidf = TfidfVectorizer(max_features=20)  # top 20 features for readability

# Fit and transform the lemmatized text
tfidf_matrix = tfidf.fit_transform(df['lemmatized_text'])

# Convert to DataFrame for display
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf.get_feature_names_out()
)

print("TF-IDF Feature Names:")
print(list(tfidf.get_feature_names_out()))

print("\nTF-IDF Matrix (shape:", tfidf_matrix.shape, "):")
print(tfidf_df.round(4))
```

## Step 7: Full Preprocessed Summary

```python
print("=== Full Preprocessing Summary ===")
summary_df = df[['text', 'cleaned_text', 'no_stopwords_text', 'lemmatized_text', 'label', 'encoded_label']]
pd.set_option('display.max_colwidth', 60)
print(summary_df.to_string(index=False))
```

## Step 8: Save Outputs to CSV Files

```python
# Save full preprocessed dataframe
df.to_csv('preprocessed_text.csv', index=False)
print("Saved: preprocessed_text.csv")

# Save TF-IDF matrix
tfidf_df['label'] = df['label'].values
tfidf_df['encoded_label'] = df['encoded_label'].values
tfidf_df.to_csv('tfidf_representation.csv', index=False)
print("Saved: tfidf_representation.csv")

# Save label encoding mapping
label_map = pd.DataFrame({'label': le.classes_, 'encoded_value': le.transform(le.classes_)})
label_map.to_csv('label_encoding_map.csv', index=False)
print("Saved: label_encoding_map.csv")

print("\nAll outputs saved successfully!")
```
